```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('hw4_exams.csv')

print("First five rows of the dataset:")
print(df.head())

print("\nBasic information about the dataset:")
df.info()

print("\nStatistical information for scores:")
print(df[['math score', 'reading score', 'writing score']].describe())

print("\nMean scores grouped by race/ethnicity:")
grouped_means = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()
print(grouped_means)

print("\nMath scores as a DataFrame:")
math_scores_df = df[['math score']]
print(math_scores_df)

print("\nMath scores as a Series (bracket notation):")
math_scores_series = df['math score']
print(math_scores_series)

print("\nMath scores as a Series (dot notation):")
math_scores_dot = df['math score']
print(math_scores_dot)

print("\nFemales with math scores â‰¥ 90:")
filtered_df = df[(df['gender'] == 'female') & (df['math score'] >= 90)]
print(filtered_df)

print("\nBox plot for scores:")
df[['math score', 'reading score', 'writing score']].plot(kind='box', title='Box Plot of Scores')
plt.title('Box Plot of Scores')
plt.show()

print("\nAverage scores by gender:")
gender_means = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print(gender_means)
gender_means.plot(kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.show()

print("\nCount plot of race/ethnicity:")
sns.catplot(x='race/ethnicity', kind='count', data=df)
plt.title('Count of Students by Race/Ethnicity')
plt.show()

print("\nCount plot with rotated x labels:")
sns.catplot(x='race/ethnicity', kind='count', data=df)
plt.title('Count of Students by Race/Ethnicity')
plt.xticks(rotation=45)
plt.show()

print("\nScatter plot of writing vs reading scores (specific method):")
sns.scatterplot(data=df, x='writing score', y='reading score', hue='test preparation course')
plt.title('Writing vs Reading Scores by Test Prep Course')
plt.show()

print("\nScatter plot of writing vs reading scores (general method):")
df.plot.scatter(x='writing score', y='reading score', c=df['test preparation course'].apply(lambda x: 'blue' if x == 'completed' else 'orange'))
plt.title('Writing vs Reading Scores by Test Prep Course')
plt.show()

print("\nScatter plot (general method, resized):")
df.plot.scatter(x='writing score', y='reading score', c=df['test preparation course'].apply(lambda x: 'blue' if x == 'completed' else 'orange'), figsize=(10, 6))
plt.title('Writing vs Reading Scores by Test Prep Course (Resized)')
plt.show()
```
